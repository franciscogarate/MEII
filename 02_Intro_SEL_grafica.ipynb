```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def y1(x): return -x
def y2(x): return 2-x
x = np.linspace(-5,5,100)
y1 = y1(x)
y2 = y2(x)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['x+y=0','x+y=2'])
plt.title('Sistema incompatible')
plt.savefig('fig2_si.png', format='png', transparent=True)
```

```python
def y3(x): return x
def y4(x): return -x
x = np.linspace(-5,5,100)
y3 = y3(x)
y4 = y4(x)
plt.plot(x, y3)
plt.plot(x, y4)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['x+y=0','x-y=0'])
plt.title('Sistema compatible determinado')
plt.savefig('fig2_scd.png', format='png', transparent=True)
```

```python
def y5(x): return 2-x
def y6(x): return 2-x+0.07 #desplazo un poco para que sea visible en la grfica
x = np.linspace(-5,5,100)
y5 = y5(x)
y6 = y6(x)
plt.plot(x, y5)
plt.plot(x, y6)
plt.grid()
plt.xlabel('x', alpha=0.5)
plt.ylabel('y', alpha=0.5)
plt.legend(['x+y=2','x+y=2'])
plt.title('Sistema compatible indeterminado')
plt.savefig('fig2_sci.png', format='png', transparent=True)
```
