## Formas cuadr√°ticas

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  

def fun(x, y):
    return  x**2 + y**2

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-3, 3, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)
#ax.set_xlabel('X Label')
#ax.set_ylabel('Y Label')
#ax.set_zlabel('Z Label')
plt.ylim(-4, 4)
plt.xlim(-4, 4)

plt.show()
#plt.savefig('fig4_subespacio2.png', format='png', transparent=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n
def fun(x, y):
    return x**2 + y**2

# Generar un grid de valores x, y
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Calcular los valores de Z para la funci√≥n
Z = fun(X, Y)

# Calcular los gradientes
Gx, Gy = np.gradient(Z, x, y)

# Crear la figura
plt.figure(figsize=(8, 6))

# Graficar la funci√≥n como un mapa de contornos
plt.contourf(X, Y, Z, levels=50, cmap='RdYlBu')
plt.colorbar()  # Mostrar la barra de colores

# Graficar los gradientes
plt.quiver(X, Y, Gx, Gy, color='k')

# Configurar los t√≠tulos y etiquetas
plt.title('Mapa de contornos y gradientes de $f(x, y) = x^2 - y^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')

# Mostrar el gr√°fico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crea una malla de puntos para evaluar el gradiente
x, y = np.meshgrid(np.linspace(-10, 10, 20), np.linspace(-10, 10, 20))

# Calcula las derivadas parciales, es decir, los componentes del gradiente
u = 2 * x  # Derivada parcial con respecto a x
v = 2 * y  # Derivada parcial con respecto a y

# Crea la figura y el eje
plt.figure(figsize=(8, 6))
plt.title('Campo de gradientes de f(x, y) = $x^2 - y^2$')

# Dibuja el campo de vectores (gradientes) con flechas
plt.quiver(x, y, u, v, color='blue')

# Establece l√≠mites para x e y para una mejor visualizaci√≥n
plt.xlim([-10, 10])
plt.ylim([-10, 10])

# Etiquetas y mostrando el gr√°fico
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  

def fun(x, y):
    return  - x**2 - y**2

def fun2(x, y):
    return  - 2*x - 2*y

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-3, 3, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun(np.ravel(X), np.ravel(Y)))
zd = np.array(fun2(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)
Zd = zd.reshape(X.shape)



ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Zd, alpha=0.7)

#ax.set_xlabel('X Label')
#ax.set_ylabel('Y Label')
#ax.set_zlabel('Z Label')
plt.ylim(-4, 4)
plt.xlim(-4, 4)

#plt.show()
#plt.savefig('fig4_subespacio2.png', format='png', transparent=True)
```
